```julia
using Revise
using MajoranaPropagation
using PauliPropagation

using Plots 
using LaTeXStrings
using TimerOutputs
using ProgressMeter
```

```julia
nq = 10 
nf = nq + 1 
is_spinful = false 

qubit_filter, fermion_filter, max_single_filter = create_filters(nq, nf, is_spinful)

hsum = hybridSum(nq, [:X, :Y], [1, 2], nf, :nn, [1, 3])
```

```julia
hsum2 = hybridSum(nq, [:Z, :Y], [2, 4], nf, is_spinful)
```

```julia
circ = []
thetas = []

for i=1:nq
    push!(circ, HybridGate([:Z], [i], :hop, [i, i+1]))
    push!(thetas, 0.1) 
end
```

```julia
propagate!(circ, hsum, thetas, qubit_filter, fermion_filter; min_abs_coeff=1e-6)
```

```julia
propagate!(circ, hsum2, thetas, qubit_filter, fermion_filter; min_abs_coeff=1e-6)
```
